### Extracting each field for preprocessing

```python
city = array(train_df['city'])
service_name = array(train_df['service_name'])
week_of_month = array(train_df['week_of_month'])
month = array(train_df['month'])
year = array(train_df['year'])
count= array(train_df['count'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(strip_accents='ascii',lowercase=False,analyzer='word',decode_error='strict')
```

### One Hot Encoding For Categorical Variables

```python
city_code = vectorizer.fit_transform(city)
service_name_code = vectorizer.fit_transform(service_name)
```

```python
import numpy as np 
data_np = np.concatenate([city_code.toarray(),service_name_code.toarray()], axis=1)
```

```python
week_of_month_arr =np.array(week_of_month).reshape(-1,1)
month_code_arr =np.array(month).reshape(-1,1)
year_arr =np.array(year).reshape(-1,1)
count_arr=np.array(count).reshape(-1,1)
```

```python
data_np = np.concatenate([city_code.toarray(),service_name_code.toarray() ,week_of_month_arr, month_code_arr ,year_arr,count_arr], axis=1)
```

```python
train= data_np[0:3708]
```

```python
class_label = train[:, -1]
dataset = train[:, :-1]
```

```python
test= data_np[3708:3773]
test_dataset = test[:, :-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset, class_label, test_size=0.20, random_state=42)
```

```python

rf_reg = RandomForestRegressor(random_state=0 ,n_estimators =2000)
rf_reg.fit(X_train, y_train)
y_pred = rf_reg.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

### Printing Actual Vs Predicted 

```python
for i,j in zip(y_test,y_pred):
    print (i,'<--------->',j)
```
